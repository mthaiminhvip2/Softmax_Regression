```python
# tạo hàm tính A(các giá trị xác suất phụ thuộc vào Z)
def softmax(Z):
  return np.exp(Z) / np.sum(np.exp(Z) , axis = 0 , keepdims= True)

# hàm tối ưu của softmax : trừ mỗi giá trị Z đi 1 lượng (c = giá trị lớn nhất trong Z) .
def softmax_stable(Z):
  c = np.max(Z , axis = 0)
  return softmax(Z-c)
```

```python
 # hàm cross entropy sau khi đã lấy log 2 vế
def cost(X,Y,W):
  Z = np.dot(W.T , X)
  A = softmax(Z);
  return -np.sum(Y * np.lof(A));
```
